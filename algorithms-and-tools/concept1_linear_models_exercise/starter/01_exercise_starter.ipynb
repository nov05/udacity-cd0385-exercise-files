```python
import numpy as np # type: ignore
import pandas as pd # type: ignore
from sklearn import datasets # type: ignore
from sklearn.model_selection import train_test_split # type: ignore
from sklearn.linear_model import LogisticRegression, SGDRegressor # type: ignore
```

## Linear Classifier

```python
# Load in the wine dataset
from sklearn.datasets import load_wine # type: ignore
wine = load_wine()
```

```python
# Create the wine `data` dataset as a dataframe and name the columns with `feature_names`
df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
df['target'] = wine.target
```

```python
# Check your dataframe by `.head()`
df.head()
```

```python
df.describe()
```

```python
# Split your data with these ratios: train: 0.8 | test: 0.2
df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
# How does the model perform on the training dataset and default model parameters?
# Using the hyperparameters in the requirements, is there improvement?
# Remember we use the test dataset to score the model
clf = (
    LogisticRegression(random_state=0, max_iter=10000)
    .fit(df_train.drop(columns=['target']), df_train['target'])
)
clf.score(df_test.drop(columns=['target']), df_test['target'])  
## default, 0.9722222222222222
## max_iter=10000, 1.0
```

## Linear Regression

```python
# Load in the diabetes dataset
from sklearn.datasets import load_diabetes   # type: ignore
diabetes = load_diabetes()
```

```python
# Create the diabetes `data` dataset as a dataframe and name the columns with `feature_names`
dfd = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names) # type: ignore

# Include the target as well
dfd['target'] = diabetes.target
```

```python
# Check your dataframe by `.head()`
dfd.head()
```

```python
# Split your data with these ratios: train: 0.8 | test: 0.2
dfd_train, dfd_test = train_test_split(dfd, test_size=0.2, random_state=42)
```

```python
# How does the model perform on the training dataset and default model parameters?
# Using the hyperparameters in the requirements, is there improvement?
# Remember we use the test dataset to score the model
reg = (
    SGDRegressor(random_state=0, max_iter=10000)
    .fit(dfd_train.drop(columns=['target']), dfd_train['target'])
)
reg.score(dfd_test.drop(columns=['target']), dfd_test['target'])  
## default, 0.4126478974893697
## max_iter=10000, 0.4598078582608498
```
